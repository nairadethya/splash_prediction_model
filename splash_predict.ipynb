```python
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras import Sequential
from tensorflow.keras import optimizers
from tensorflow.keras import layers
from keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import array_to_img, img_to_array, load_img

import numpy as np 
import pandas as pd 

import matplotlib.pyplot as plt 
import matplotlib.image as mpimg

import cv2
from google.colab.patches import cv2_imshow

import os
```

```python
train_img = cv2.imread('/content/ns_1.tif', 0)
train_img = cv2.resize(train_img, (250, 250))
train_img = train_img.reshape(1, -1)
train_img = train_img/255
test_img = cv2.imread('/content/drive/MyDrive/drop_test/s_t_1.tif', 0)
test_img = cv2.resize(test_img, (250, 250))
test_img = test_img.reshape(1, -1)
test_img = test_img/255
print(train_img.shape)
print(test_img.shape)
```

```python
#Defining the model
model = keras.Sequential(
    keras.layers.Dense(2, activation = 'sigmoid', input_shape = [250*250])
    )
```

```python
#Loading the training images

for i in range(2,701):
  img = cv2.imread('/content/ns_'+str(i)+'.tif', 0)
  img = cv2.resize(img, (250, 250))
  img = img/255
  img = img.reshape(1,-1)
  train_img = np.append(train_img, img, axis = 0)
  

print(train_img.shape)

train_img = tf.stack(train_img)
```

```python
#Loading the testing images
for i in range(2, 34):
  t_img = cv2.imread('/content/drive/MyDrive/drop_test/s_t_'+str(i)+'.tif', 0)
  t_img = cv2.resize(img, (250, 250))
  t_img = t_img/255
  t_img = t_img.reshape(1, -1)
  test_img = np.vstack((test_img, t_img))


test_img = tf.stack(test_img)
print(test_img.shape)
```

```python
#Loading the labels as the previous ones where wrong
train_labels = pd.read_csv('/content/train_labels _updated.csv', header = 0)
test_labels = pd.read_csv('/content/drive/MyDrive/test_labels_new.csv', header = 0)

train_labels = train_labels.to_numpy()
test_labels =  test_labels.to_numpy()

train_labels = tf.stack(train_labels)
test_labels = tf.stack(test_labels)

print(train_labels.shape)
print(test_labels.shape)
```

```python
epoch = 10000
batch = 32
split = 0.2
LR = 0.0005
Mmntm = 0.9
ptnc = 50

callback = [keras.callbacks.EarlyStopping(
    monitor='loss', min_delta=0,
    patience=ptnc, restore_best_weights=True)]

opt = optimizers.SGD(learning_rate = LR, momentum = Mmntm)

model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_img, train_labels, batch_size = batch, epochs = epoch, verbose = 2, validation_split = split, callbacks = callback)
```

```python
loss, acc = model.evaluate(test_img, test_labels)
print('test accuracy:', acc*100, 'loss:', loss)
```

```python
predictions = model.predict(test_img)
```
